```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# Data Prep

```python
data=pd.read_csv('BreastCancerData.csv')
```

```python
data.head()
```

```python
data=data.drop(['id'],axis=1)
```

```python
data.head()
```

```python
def myfunction(x):
    if x == 'B':
        return 0
    elif x == 'M':
        return 1
    
data['diagnosis']=data['diagnosis'].apply(myfunction)
data.head()
```

```python
#standardise data
from sklearn.preprocessing import StandardScaler
X=StandardScaler().fit_transform(data.iloc[:,1:])
X.shape
```

# Determining a suitable epsilon & minPts

We find a suitable value for epsilon by calculating the distance to the nearest n points for each point, sorting and plotting the results. Then we look to see where the change is most pronounced (greatest change in gradient) and select that as epsilon.

We can calculate the distance from each point to its closest neighbour using the NearestNeighbors. The point itself is included in n_neighbors.

```python
from sklearn.neighbors import NearestNeighbors

neigh=NearestNeighbors(n_neighbors=2)
nbrs=neigh.fit(X)
distances, indices = nbrs.kneighbors(X)
```

```python
distances=np.sort(distances, axis=0)
distances=distances[:,1]
plt.plot(distances)
```

From the above graph we will take eps to be 4.

Determining number of minimum points with heuristical technique:
- minPts = ln(n) - where n is the total number of points to be clustered.
- Therefore minPts = ln(569) ≈ 6 (To nearest whole number)

# Without PCA

```python
from sklearn.cluster import DBSCAN
```

```python
import time
#start timer
start_time=time.time()
dbscan=DBSCAN(eps=4,min_samples=6)
clusters=dbscan.fit_predict(X)
#stop timer
end_time=time.time()
duration=end_time-start_time
#printing duration of timing
print("Duration of model training: %s"%(duration))
```

```python
colors='rgbkcmy'
```

```python
np.unique(clusters)
```

This means, there's only one cluster and outliers. The model has performed very poorly. Tried with 7,12 and 60 minPts and still gave same number of results.

```python
from sklearn import preprocessing
j=preprocessing.MinMaxScaler().fit_transform(X)
j.shape
```

```python
for i in np.unique(clusters):
    label = "Outlier" if i==-1 else 'Cluster '+str(i+1)
    plt.scatter(j[clusters==i,1],j[clusters==i,2],
                color=colors[i],label=label)
    
plt.xlabel(data.columns[1])
plt.ylabel(data.columns[2])

plt.legend()
plt.title('DBScan Clustering')
plt.show()
```

# With Feature Engineering

```python
df=data.iloc[:,:]
df.shape
```

```python
df_std=(df-df.mean())/(df.std())
```

```python
df_std.head()
```

```python
corr_mat=df_std.corr()
```

```python
plt.figure(figsize=(25,10))
sns.heatmap(data=corr_mat,annot=True,cmap='GnBu')
plt.title('Correlation Matrix for Breast Cancer Wisconsin (Diagnostic) Dataset')
plt.show()
```

```python
target='diagnosis'
candidates=corr_mat.index[(corr_mat[target]>0.5)| (corr_mat[target]<-0.5)].values
```

```python
candidates = candidates[candidates!=target]
print('Correlated to',target,', ',candidates)
```

```python
removed=[]
for c1 in candidates:
    for c2 in candidates:
        if (c1 not in removed) and (c2 not in removed):
            if c1!=c2:
                coeff = corr_mat.loc[c1,c2]
                if coeff>0.6 or coeff<-0.6:
                    removed.append(c1)
print('Removed: ',removed)
selected_features=[x for x in candidates if x not in removed]
print('Selected features: ', selected_features)
```

```python
xf = df_std.drop(removed,axis=1)
```

```python
xf.head()
```

```python
xf.shape
```

```python
xf2=xf.iloc[:,1:17].values
```

```python
xf2.shape
```

```python
start_time=time.time()
dbscan=DBSCAN(eps=4,min_samples=6)
clusters=dbscan.fit_predict(xf2)
#stop timer
end_time=time.time()
duration=end_time-start_time
#printing duration of timing
print("Duration of model training: %s"%(duration))
```

```python
np.unique(clusters)
```

Lesser time as well. Same poor results.

```python
m=preprocessing.MinMaxScaler().fit_transform(xf2)
m.shape
```

```python
for i in np.unique(clusters):
    label = "Outlier" if i==-1 else 'Cluster '+str(i+1)
    plt.scatter(m[clusters==i,1],m[clusters==i,2],
                color=colors[i],label=label)
    
plt.xlabel(xf.columns[1])
plt.ylabel(xf.columns[2])

plt.legend()
plt.title('DBScan Clustering (With Feature Engineering)')
plt.show()
```
